```python
import requests
import pandas as pd
import re
import os
from bs4 import BeautifulSoup
from tqdm import tqdm
```

## get artists main page URL

```python
url_JJ = "https://www.lyrics.com/artist/Jack-Johnson/468749"
url_DB = "https://www.lyrics.com/artist/David-Bowie/3753"
```

```python
# different parts of URL for DB
main_url = "https://www.lyrics.com"
artist_url = "artist"
artist_name = "David-Bowie"
artist_code = 3753
full_url = f'{main_url}/{artist_url}/{artist_name}/{artist_code}'
full_url
```

```python
# Get the page and convert to Beautiful soup
soup = BeautifulSoup(requests.get(full_url).text)
```

```python
hrefs = soup.find_all('a')
```

```python
hrefs
```

```python
# Get all hrefs
DB_list = []
hrefs = soup.find_all('a')
for href in hrefs:
    my_href = href["href"]
    if "lyric/" in my_href:
        DB_list.append(f'{main_url}{my_href}')
```

```python
DB_list
```

```python
save_folder = "../week_4/David_Bowie/Text" 
os.makedirs(save_folder, exist_ok=True)  

for link in tqdm(DB_list, desc="Downloading Lyrics"):
    try:
        soup = BeautifulSoup(requests.get(link).text)
        lyrics = soup.find("pre").text

        # Extract the link name
        link_name = link.split('/')[-1]
        # Create the file name by adding '.txt' extension
        file_name = os.path.join(save_folder, link_name + '.txt')
        
        # Save the lyrics in a separate text file
        with open(file_name, 'w', encoding='utf-8') as file:
            file.write(lyrics)
    except AttributeError:
        continue
```

```python
#create a list from the single files in the folder
save_folder = "../week_4/David_Bowie/Text" 

song_list = os.listdir(save_folder)
song_list
```

```python
len(song_list)
#lots of duplicates 
```

```python
# some songs are double, found a way to elimnate: fuzzywuzzy
#apply the same treshold as for JJ
from fuzzywuzzy import fuzz
song_list
new_song_list=[]
threshold = 58

for song in song_list:
    is_duplicate = False
    for unique_song in new_song_list:
        similarity_score = fuzz.token_sort_ratio(song, unique_song)
        if similarity_score >= threshold:
            is_duplicate = True
            break
    if not is_duplicate:
        new_song_list.append(song)
        
new_song_list
```

```python
len(song_list), len(new_song_list)
```

```python
#create a cropus with all the songs in the new_song_list 
folder_path = "../week_4/David_Bowie/Text"
corpus = []
artist=[]

for song in new_song_list:
    file_name = song
    file_path = os.path.join(folder_path, file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        lyrics = file.read()
        corpus.append(lyrics)
        artist.append("David Bowie")

corpus
```
